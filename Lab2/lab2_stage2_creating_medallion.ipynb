Creating the medallion architecture

```python
dbutils.widgets.text("catalog_name", "dbr_dev")
catalog_name = dbutils.widgets.get("catalog_name")

container = 'trezio2005'
storage_account = 'dlspl21databricks'
abfss_path = f'abfss://{container}@{storage_account}.dfs.core.windows.net/'

#creating the medallion architecture
spark.sql(f"""
            CREATE SCHEMA IF NOT EXISTS {catalog_name}.trezio2005_bronze
            MANAGED LOCATION '{abfss_path}bronze'
          """)

spark.sql(f"""
            CREATE SCHEMA IF NOT EXISTS {catalog_name}.trezio2005_silver
            MANAGED LOCATION '{abfss_path}silver'
          """)

spark.sql(f"""
            CREATE SCHEMA IF NOT EXISTS {catalog_name}.trezio2005_gold
            MANAGED LOCATION '{abfss_path}gold'
          """)
```
